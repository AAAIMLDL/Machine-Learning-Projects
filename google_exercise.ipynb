```python
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth":2.5})
# sns.displot(googledata, x = "Rating", hue="Genres", kind="kde")
sns.displot(googledata, x = "Rating", hue="Category", kind="kde")
```
