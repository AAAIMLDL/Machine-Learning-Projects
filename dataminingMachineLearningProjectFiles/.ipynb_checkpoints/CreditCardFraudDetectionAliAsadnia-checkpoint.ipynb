Class imbalance occurs when the number of instances in one class significantly outnumbers the instances in the other class. In this dataset, the `Class` column, which we are using as the target variable for a binary classification task, exhibits a notable imbalance. Specifically, the dataset contains 284,315 instances of class `0` and only 492 instances of class `1`.

To mitigate the effects of class imbalance, several strategies can be employed:
1. **Resampling Techniques**:
   - **Oversampling the Minority Class**: Techniques such as SMOTE (Synthetic Minority Over-sampling Technique) can generate synthetic samples for the minority class.
   - **Undersampling the Majority Class**: Reducing the number of instances in the majority class to balance the dataset.
2. **Algorithmic Approaches**:
   - **Using Ensemble Methods**: Techniques such as balanced random forests or boosting methods that focus on the minority class.
3. **Evaluation Metrics**:
   - **Instead of relying solely on accuracy, we can use metrics such as precision, recall, F1-score, ROC-AUC, and confusion matrix to get a better understanding of the model's performance on both classes.**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import classification_report, roc_auc_score

# Define features and target
X_robust = df[['TimeScaledRobust', 'AmountScaledRobust']]
X_minmax = df[['scaled_timeMinMax', 'scaled_amountMinMax']]
y = df['Class']  # Assuming 'Class' is your target variable

# Split data into training and testing sets
X_train_robust, X_test_robust, y_train, y_test = train_test_split(X_robust, y, test_size=0.3, random_state=42)
X_train_minmax, X_test_minmax, y_train, y_test = train_test_split(X_minmax, y, test_size=0.3, random_state=42)

# Initialize the RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train and evaluate the model using RobustScaler data
clf.fit(X_train_robust, y_train)
y_pred_robust = clf.predict(X_test_robust)
roc_auc_robust = roc_auc_score(y_test, clf.predict_proba(X_test_robust)[:, 1])
print("RobustScaler Classification Report:\n", classification_report(y_test, y_pred_robust))
print("RobustScaler ROC-AUC Score:", roc_auc_robust)

# Train and evaluate the model using MinMaxScaler data
clf.fit(X_train_minmax, y_train)
y_pred_minmax = clf.predict(X_test_minmax)
roc_auc_minmax = roc_auc_score(y_test, clf.predict_proba(X_test_minmax)[:, 1])
print("MinMaxScaler Classification Report:\n", classification_report(y_test, y_pred_minmax))
print("MinMaxScaler ROC-AUC Score:", roc_auc_minmax)
```

```python
# Perform PCA to reduce dimensions
pca = PCA(n_components=0.95)
train_features_pca = pca.fit_transform(train_features)
test_features_pca = pca.transform(test_features)

# Perform a more extensive grid search for DBSCAN parameters
param_grid = {
    'eps': np.arange(0.1, 1.1, 0.05),
    'min_samples': range(3, 20)
}

best_score = -1
best_params = {}

for eps in param_grid['eps']:
    for min_samples in param_grid['min_samples']:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        clusters = dbscan.fit_predict(train_features_pca)
        if len(set(clusters)) > 1:
            score = silhouette_score(train_features_pca, clusters)
            if score > best_score:
                best_score = score
                best_params = {'eps': eps, 'min_samples': min_samples}

print(f"Best parameters: {best_params} with silhouette score: {best_score}")
```

```python
# Initialize and fit the DBSCAN model with specified parameters
dbscan = DBSCAN(eps=0.1, min_samples=5)
dbscan.fit(train_features)
```

```python
# Predict clusters on the training set
train_clusters = dbscan.labels_
train_anomalies = (train_clusters == -1).astype(int)
```

```python
# Predict clusters on the test set
test_clusters = dbscan.fit_predict(test_features)
test_anomalies = (test_clusters == -1).astype(int)
```

```python
# Evaluate the model
print("Classification Report:\n", classification_report(test_labels, test_anomalies))
```

### Local Outlier Factor

```python
# Initialize the LOF model
lof = LocalOutlierFactor(n_neighbors=40, algorithm='auto', leaf_size=30, metric='minkowski',p=2, 
                         metric_params=None, contamination=float(0.05), novelty=True)
# Fit the model on the training data
lof.fit(train_features)
# Predict the anomalies on the test data
y_pred = lof.predict(test_features)
# Convert predictions: LOF returns -1 for outliers, 1 for inliers. Convert to 1 for anomalies, 0 for normal.
y_pred = np.where(y_pred == -1, 1, 0)
```

```python
# Evaluate the model
print("Classification Report:\n", classification_report(test_labels, y_pred))
print("Accuracy Score:", accuracy_score(test_labels, y_pred))
```

```python
# Add the predictions to the test_features for visualization
test_data_viz = pd.DataFrame(test_features, columns=['Feature' + str(i) for i in range(test_features.shape[1])])
test_data_viz['Anomaly'] = y_pred
```
